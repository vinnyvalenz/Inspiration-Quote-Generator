Here is where I used Max Woolf method of training GPT-2 on new data, for the purpose of it on the motivational quote dataset


```
file_name = "Motivational-Quotes-Database.txt"
```

```
gpt2.copy_file_from_gdrive(file_name)
```

```
sess = gpt2.start_tf_sess(). 

gpt2.finetune(sess,
              dataset=file_name,
              model_name='355M',
              steps=1000,
              restore_from='fresh',
              overwrite = True,
              run_name='run3',
              print_every=10,
              sample_every=200,
              save_every=500,
              )
```
